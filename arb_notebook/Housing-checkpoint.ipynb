# Apt/Housing For Rent

```python
#================================================================================
# PARAMETERS
#================================================================================
LOCALE = "denver"  # Available: devner, oakland



#================================================================================
%pylab inline
%matplotlib inline

import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reloading data
_df = pd.read_csv('data/' + LOCALE + '--house.csv')
df = _df

# Remove: price is null
df = df[df['price'].notnull()]
# Remove: price > $20,000
df = df[df['price'] <= 20000]
# Remove: price < $100
df = df[df['price'] >= 100]
# Remove: bedroom number is null
df = df[df['num_bd'].notnull()]

# saving temporary data
#df.to_csv('data/' + LOCALE + '--house_edited.csv', index=False)
```

## Prices

```python
fig = plt.figure(figsize=(10,7)) # in inches
ax = fig.add_subplot(111)

# Filters
__df = df
__df = __df[__df['housing_type']=='apartment']
__df = __df[__df['num_bd']==2]


# data
x = __df['price']
max_price = max(__df['price'])
# bins = [0, 500, 1000, 1500, 2000, 2500]
bins = range(0, 10000, 100)

# Chart: -------------------------------------
# aesthetics
ax.set_title("Distribution By Prices", fontsize=20)
ax.set_xlabel('Prices', fontsize=16)
ax.set_ylabel('Counts', fontsize=16)
aes = {
    "edgecolor": "#ffffff",
    "color": "#008AB8"
}
# histogram
ax.hist(x, bins=bins, alpha=0.8, **aes)
plt.show()
```

### Houseing Type

```python
_types = list(set(df['housing_type']))
types = []
counts = []
for t in _types:
    if pd.isnull(t):
        count = sum(df['housing_type'].isnull())
        counts.append(count)
        types.append('unknown')
    else:
        count = sum(df['housing_type'] == t)
        counts.append(count)
        types.append(t)

x = types
y = counts

# Chart: ------------------------------------------------------
fig = plt.figure(figsize=(10,7)) # canvas size in inches
ax = fig.add_subplot(111)

# ax.set_title("XXX", fontsize=20)
ax.set_xlabel('Housing Types', fontsize=16)
ax.set_ylabel('Unit Count', fontsize=16)
aes = {
    "edgecolor": "#ffffff",
    "color": "#008AB8"
}

plt.bar(range(len(x)), y, align='center', **aes)
plt.xticks(range(len(x)), x)
plt.show()
```

### Number of Bedroom

```python
__df = df[df['housing_type']=='house']
group_by = __df.groupby('num_bd')
groups = group_by.groups
group_keys = []
group_counts = []
for k in groups.keys():
    key = str(int(k)) + " bd"
    count = len(group_by.get_group(k))
    group_keys.append(key)
    group_counts.append(count)

x = group_keys
y = group_counts

# Chart: ------------------------------------------------------
fig = plt.figure(figsize=(10,7)) # canvas size in inches
ax = fig.add_subplot(111)

# ax.set_title("XXX", fontsize=20)
ax.set_xlabel('Number of Bedroom', fontsize=16)
ax.set_ylabel('Unit Count', fontsize=16)
aes = {
    "edgecolor": "#ffffff",
    "color": "#008AB8"
}

plt.bar(range(len(x)), y, align='center', **aes)
plt.xticks(range(len(x)), x)
plt.show()
```

### Square footage

```python
fig = plt.figure(figsize=(10,7)) # in inches
ax = fig.add_subplot(111)

# data
x = df['total_sq_ft']
max_sq_ft = max(df['total_sq_ft'])
# bins = [0, 500, 1000, 1500, 2000, 2500]
bins = range(0, 5000, 200)

# Chart: -------------------------------------
# aesthetics
ax.set_title("Distribution By Square Footage", fontsize=20)
ax.set_xlabel('Square Feet', fontsize=16)
ax.set_ylabel('Unit Counts', fontsize=16)
aes = {
    "edgecolor": "#ffffff",
    "color": "#008AB8"
}
# histogram
ax.hist(x, bins=bins, alpha=0.8, **aes)
plt.show()
```

### Number of Bathroom

```python
group_by = df.groupby('num_ba')
groups = group_by.groups
group_keys = []
group_counts = []
for k in groups.keys():
    key = str(int(k)) + " ba"
    count = len(group_by.get_group(k))
    group_keys.append(key)
    group_counts.append(count)

# count the null
group_keys.append("unknown")
group_counts.append(sum(df['num_ba'].isnull()))

x = group_keys
y = group_counts

# Chart: ------------------------------------------------------
fig = plt.figure(figsize=(10,7)) # canvas size in inches
ax = fig.add_subplot(111)

# ax.set_title("XXX", fontsize=20)
ax.set_xlabel('Number of Bathroom', fontsize=16)
ax.set_ylabel('Unit Count', fontsize=16)
aes = {
    "edgecolor": "#ffffff",
    "color": "#008AB8"
}

plt.bar(range(len(x)), y, align='center', **aes)
plt.xticks(range(len(x)), x)
plt.show()
```

### Price Per Square Foot By Apt Type

```python
# data
__df = df
__df = __df[__df['price'].notnull()]
__df = __df[__df['total_sq_ft'].notnull()]

group_by = __df.groupby('apt_type')
groups = group_by.groups
keys = []
sq_feet_prices = []

_keys = groups.keys()
for k in _keys:
    key = k
    group = group_by.get_group(k)
    x = group['price'] / group['total_sq_ft']
    p = np.mean(x)
    
    keys.append(key)
    sq_feet_prices.append(p)

x = keys
y = sq_feet_prices
data = zip(x, y)
x = [z[0] for z in sorted(data, key=lambda x: x[1], reverse=True)]
y = [z[1] for z in sorted(data, key=lambda x: x[1], reverse=True)]

# Chart: ------------------------------------------------------
fig = plt.figure(figsize=(10,7)) # canvas size in inches
ax = fig.add_subplot(111)

# ax.set_title("XXX", fontsize=20)
ax.set_xlabel('Number of Bathroom', fontsize=16)
ax.set_ylabel('Average price per square foot', fontsize=16)
aes = {
    "edgecolor": "#ffffff",
    "color": "#008AB8"
}

plt.bar(range(len(x)), y, align='center', **aes)
plt.xticks(range(len(x)), x)
plt.show()
```
